### 读取数据

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
from sklearn import datasets
df=pd.DataFrame(datasets.load_iris()['data'],columns=datasets.load_iris()['feature_names']) 
```

#### 数据可视化

```python
df.plot.scatter(x=0,y=1)
```

```python
df.plot.scatter(x=2,y=3)
```

#### 设定k值为3进行训练

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3).fit(df)
```

#### 输出所有样本点的聚类结果

```python
kmeans.predict(df)  #前三行预测为0类,后三行预测为1类
```

#### 3个类别的类均值

```python
kmeans.cluster_centers_ 
```

#### 以花萼长度(sepal length),花瓣长度(petal length)为x,y轴,可视化聚类结果

```python
y=kmeans.predict(df)
```

```python
sns.scatterplot(x=df['sepal length (cm)'].values,y=df['petal length (cm)'].values,hue=y)
```
